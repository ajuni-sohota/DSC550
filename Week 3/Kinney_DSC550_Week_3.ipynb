```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
plt.rcdefaults()
%matplotlib inline

gapminder = pd.read_csv("data/gapminder-FiveYearData.csv")
gapminder.head()
```

```python
# horizontal bar chart
life_exp = gapminder.groupby('continent')['lifeExp'].mean()
life_exp.plot(color='purple', kind='bar')
plt.xlabel('Continent',)
plt.ylabel('Avg Life Expectancy')
plt.title('Average Life Expectancy by Continent', color='green')
plt.legend()
plt.show()
```

```python
# line


life_exp.plot(color='blue', kind='line', 
             path_effects=[path_effects.SimpleLineShadow(),
                       path_effects.Normal()])
plt.xlabel('Continent')
plt.ylabel('Avg Life Expectancy')
plt.title('Average Life Expectancy by Continent')
plt.legend()
plt.show()
```

```python
# box and whiskers
df_lifeexp = gapminder[['continent', 'lifeExp']]
df_lifeexp.boxplot(column='lifeExp', by='continent')
plt.xlabel('Continent')
plt.ylabel('Avg Life Expectancy')
plt.title('\nAverage Life Expectancy by Continent')
plt.show()
```

### Ranking my three charts

Based on the three charts above, all plotting the same data, I rank them most effective to least effective as follows:

1. Box and whiskers (chart 3)
2. Bar chart (chart 1)
3. Line chart (chart 2)

I will explain my reasoning in reverse order.

As we're analyzing categorical data and not numerical, the line chart, while not overtly misleading, does not provide a concrete representation of the data. For instance, there is no "mystery continent" between Africa and the Americas with an average life expectancy of 57 years. Therefore, line charts should be avoided for this type of visualization.

The bar chart is good as far as it goes. It's great for initial data exploration, or if the audience just wants a high-level visualization of average life expectancy. The danger is that, if the data has not been pre-processed to account for missing data, NaN's and outliers, the simplicity of the bar chart can very easily present an inaccurate and misleading picture.

The box and whiskers clearly presents the most information and most accurate representation of the data:  
* Minimum value: The smallest value in the data set
* Second quartile: The value below which the lower 25% of the data are contained
* Median value: The middle number in a range of numbers
* Third quartile: The value above which the upper 25% of the data are contained
* Maximum value: The largest value in the data set (2)

I can see the data point gaps at the low end for the continent of Africa. However for me the strength of the box and whickers plot is that you can immediately see if you have outliers. Right up front I can go analyze those data points to ascertain whether or not they are valid and should be included in further analysis, or they should be modified or removed.

Lastly I wanted to touch on why I did not choose a pie chart. I've picked up on the fact that pie charts are lepers in the Data Science community. And, apparently for good reason.
```python
(ggplot(gapminder, aes(x='continent', y='lifeExp', fill='continent')) +
        geom_boxplot() +
        labs(title="Life Expectancy by Continent",
                     x="Continent",
                     y="Life Expectancy",
                     color="Continent")
)
```

```python
(ggplot(gapminder, aes(x='continent', y='lifeExp', fill='continent')) +
    geom_bin2d(stat='bin_2d', position='identity') +
     labs(title="Life Expectancy by Continent",
                     x="Continent",
                     y="Life Expectancy",
                     color="Continent")
)
```

```python
(ggplot(gapminder, aes(x='year', y='lifeExp', group='country', color='continent')) +
        geom_line(alpha = 0.5)
)
```
